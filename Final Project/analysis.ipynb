## 1. data preparation

### 1a. load and filter to prcdda

```python
DEMO_PATH = "DemoStats_2025_CSV_fixed/DemoStats_(DST)_2025_2025_GEO.csv"
HHS_P1    = "HouseholdSpend_2025_CSV_fixed/HouseholdSpend_(HHS)_2025_GEO_P1.csv"
HHS_P2    = "HouseholdSpend_2025_CSV_fixed/HouseholdSpend_(HHS)_2025_GEO_P2.csv"

def load_filtered(path, geo_level="PRCDDA", chunksize=100_000):
    """read csv in chunks, keep only the target geo level."""
    chunks = []
    for chunk in pd.read_csv(path, chunksize=chunksize, low_memory=False):
        geo_col = "GEO" if "GEO" in chunk.columns else None
        if geo_col is None:
            chunks.append(chunk)
        else:
            chunks.append(chunk[chunk[geo_col] == geo_level])
    return pd.concat(chunks, ignore_index=True)

print("loading demostats...")
demo = load_filtered(DEMO_PATH)
print(f"demostats shape: {demo.shape}")

print("loading hhs p1...")
hhs1 = load_filtered(HHS_P1)
print(f"hhs p1 shape: {hhs1.shape}")

print("loading hhs p2...")
hhs2 = load_filtered(HHS_P2)
print(f"hhs p2 shape: {hhs2.shape}")

# p1 and p2 are different variables for the same areas, so merge side-by-side
hhs = pd.merge(hhs1, hhs2, on="CODE", suffixes=("_p1", "_p2"))
print(f"hhs merged shape: {hhs.shape}")

# now merge demo + hhs
df = pd.merge(demo, hhs, on="CODE", suffixes=("_demo", "_hhs"))
print(f"\nfinal merged shape: {df.shape}")
print(f"rows: {df.shape[0]}, columns: {df.shape[1]}")
```
